```python
import csv
import math

input_path = '/content/drive/MyDrive/Colab Notebooks/input.csv'
output_path = '/content/drive/MyDrive/Colab Notebooks/output.csv'
```

```python

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.isqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

ranges = []
with open(input_path, 'r') as infile:
    reader = csv.reader(infile)
    next(reader)
    for row in reader:
        start, end = map(int, row)
        ranges.append((start, end))

results = []
for start, end in ranges:
    prime_count = sum(1 for num in range(start, end + 1) if is_prime(num))
    results.append((start, end, prime_count))

with open(output_path, 'w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(['Start', 'End', 'PrimeCount'])
    writer.writerows(results)

print(f"Prime counts written to {output_path}")
```
